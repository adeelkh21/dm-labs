>Since Age is a numerical value without any major
categories therefore mean is used to fill up null values

>For Cabin and Embarked are  categorical attributes therfore we use mode to fill up null

>Since age is linear, linear interpolation could be applied over it. (but we dont have any missing values in age attribute)

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Age'] = data['Age'].interpolate(method='linear')
```

```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
#droping irrelevant columns before applying SMOTE since these columns contain non numeric and non useful info
data = data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
#label encoding:categorical columns into numerical
label_encoders = {}
categorical_columns = ['Gender', 'Embarked'] #colums being encoded

for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
# Show class distribution before balancing
print("Class distribution before balancing:")
print(data['Survived'].value_counts())
```

```python
#spliting data into features X and target y
X = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
#class distribution before balancing
plt.figure(figsize=(6, 4))JytGkY
```

```python
#SMOTE (Over-sampling)
smote = SMOTE(random_state=42)
X_resampled_smote, y_resampled_smote = smote.fit_resample(X, y)
```

```python
#class distribution after SMOTE
print(pd.Series(y_resampled_smote).value_counts())
```

```python
#class distribution after SMOTE
plt.figure(figsize=(6, 4))
sns.countplot(x=y_resampled_smote)
plt.title('Class Distribution After SMOTE')
plt.show()
```

```python
#Random Undersampling
undersample = RandomUnderSampler(random_state=42)
X_resampled_undersample, y_resampled_undersample = undersample.fit_resample(X,y)
```

```python
#class distribution after Random Undersampling
print(pd.Series(y_resampled_undersample).value_counts())
```

```python
#class distribution after Random Undersampling
plt.figure(figsize=(6, 4))
sns.countplot(x=y_resampled_undersample)
plt.title('Class Distribution After Random Undersampling')
plt.show()
```
