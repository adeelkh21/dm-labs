# **TASK 1**

```python
dictionary = {"perfume":1 , "studs":12, "shirt":15, "shorts":9 ,"belt":7 }
```

```python
#display invevtory
def display_products():
    for product, quantity in dictionary.items():
        print(f"{product}: {quantity}")
```

```python
#sell product
def sell_product(product, quantity):
    if product in dictionary:
        if dictionary[product]>=quantity:
            dictionary[product]=dictionary[product]-quantity
            print(f"\n{quantity} {product} sold")
        else:
            print("\n stock insufficient")
    else:
        print("\n product not found!")
```

```python
#restock products
def restock_product(product, quantity):
    if product in dictionary:
        dictionary[product]=dictionary[product]+quantity
        print(f"\n{quantity} {product} restocked")
    else:
        print("\n product not found!")
```

```python
display_products()
sell_product("belt",2)
sell_product("perfume",1)
sell_product("shirt",5)
display_products()
restock_product("perfume",5)
display_products()
```
